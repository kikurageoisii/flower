````python
# 必要ライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
````

## Irisデータセットの読み込みと確認

![slide45](slide/スライド45.jpg)

````python
dataset = datasets.load_iris()    # irisデータセットの読み込み
````

````python
# データの取り出し方を確認
print("targets:")
print(dataset.target, "\n")    # ラベルの表示
print("target_names:")
print(dataset.target_names, "\n")    # ラベル番号とアヤメの種類の対応を確認
print("data:")
print(dataset.data[:5], "\n")    # データの表示(最初の5つのみ)
print("feature_names:")
print(dataset.feature_names, "\n")    # データの各列が示す特徴の確認
````

![slide46](slide/スライド46.jpg)

![slide47](slide/スライド47.jpg)

````python
#データをプロットするための関数
def plot_all(features, target_names, data, clusters, print_label=False):
    
    fig = plt.figure(figsize=(12.8, 7.2), tight_layout=True)
    subplot = 1
    
    #すべての特徴量(変数)の組み合わせについて、グラフを表示する
    for i, feature_a in enumerate(features):
        for j, feature_b in enumerate(features[i+1:]):
            
            ax = fig.add_subplot(2, 3, subplot, xlabel=feature_a, ylabel=feature_b)
            
            #各アヤメのプロット点を取得する
            for k, name in enumerate(target_names):
                plot_a = [
                    d[i] for d, c in zip(data, clusters) if c==k
                ]
                plot_b = [
                    d[i+1+j] for d, c in zip(data, clusters) if c==k
                ]
                ax.scatter(plot_a, plot_b, label=name)
                
            ax.legend() if print_label else None
            subplot += 1
            
    plt.plot()
````

````python
features = dataset.feature_names
target_names = dataset.target_names
data = dataset.data
targets = dataset.target
plot_all(features, target_names, data, targets, print_label=True) #プロットする
````

## k-meansの実行と評価

````python
predicted_clusters = KMeans(n_clusters=3, init="random", n_init=1).fit_predict(dataset.data) #k-meanの実行
plot_all(features, target_names, data, predicted_clusters)
````

`Kmeans`クラスの概要

- 定義

```python
class sklearn.cluster.KMeans(
    n_clusters=8, *, init='k-means++', n_init=10, max_iter=300, tol=0.0001,
    precompute_distances='deprecated', verbose=0, random_state=None,
    copy_x=True, n_jobs='deprecated', algorithm='auto'
)
```

- 主要な初期化引数に関して

|引数名|概要|型|デフォルト値|
|:-|:-|:-|:-|
|`n_clusters`|形成するクラスターの数|int|8|
|`init`|k-meansの初期化手法<br>- `"k-means"`: 改良された初期化処理を実行<br>- `"random"`: 通常のk-meansを実行|str|`"k-means++"`|
|`n_init`|異なる初期値でk-meansを繰り返す回数|int|10|

- 主要なメソッドに関して

|メソッド名|概要|
|:-|:-|
|`fit(X[, y, sample_weight])`|クラスタリングを行う|
|`predict(X[, sample_weight])`|`X`の各サンプルが属する最も近いクラスターを予測する|
|`fit_predict(X[, y, sample_weight])`|`fit()`と`predict()`を両方実行する|
|`get_params([deep])`|k-meansの各パラメータを取得する|
