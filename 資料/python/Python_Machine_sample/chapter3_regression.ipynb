## データの準備

![slide29](slide/スライド29.JPG)

````python
import pandas as pd

# csvファイルに記載された内容をpandasのDataFrameオブジェクトとして読み込む
# (headerは見出し行、sepは区切り文字)
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/'
                 'housing/housing.data',
                 header=None,
                 sep='\s+')

# 各列のラベルを設定
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 
              'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
              'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

# csvファイルを保存
df.to_csv('housing.csv')
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# グラフのスタイルを指定（whitegridは白背景グリッド、notebookはデフォルトサイズ）
sns.set(style='whitegrid', context='notebook')

# 13個の説明変数のうちの5つのリスト
cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']

# 相関係数を計算
cm = np.corrcoef(df[cols].values.T)

# 全体のスケールとは別にフォントサイズを指定
sns.set(font_scale=1.5)

# 第一引数の相関係数をもとにヒートマップを作成
hm = sns.heatmap(cm,
                 cbar=True,               # カラーバーの表示
                 annot=True,              # データ値の表示
                 square=True,             # 各矩形の正方形化(縦と横のサイズを一致させる)
                 fmt='.2f',               # 数値などの表示形式
                 annot_kws={'size': 15},  # データ値のサイズの設定
                 yticklabels=cols,        # 行のメモリのラベル名
                 xticklabels=cols)        # 列のメモリのラベル名

# ヒートマップの表示
plt.show()
````

## 単回帰分析の実行と評価

`LinearRegression`クラスの概要

- 定義
```python
class sklearn.linear_model.LinearRegression(
    *, fit_intercept=True, normalize=False, 
    copy_X=True, n_jobs=None, positive=False
)
```

- 主要な初期化引数に関して

|引数名|概要|型|デフォルト値|
|:-|:-|:-|:-|
|`fit_intercept`|モデルの切片を計算するかどうか|bool|True|
|`normalize`|説明変数を正規化するかどうか|bool|False|

- 主要なメソッドに関して

|メソッド名|概要|
|:-|:-|
|`fit(X, y, sample_weight=None)`|線形回帰モデルのあてはめを実行|
|`predict(X)`|作成したモデルを利用して予測を実行|
|`get_params(deep=True)`|推定に用いたパラメータを取得|




![slide34](slide/スライド34.JPG)

````python
from sklearn.linear_model import LinearRegression

# 説明変数（1戸あたりの平均部屋数）
x = df[['RM']].values

# 目的変数 (住宅価格の中央値（単位: 1000ドル）)
y = df[['MEDV']].values

# 回帰モデルのインスタンスを作成
model_lr = LinearRegression()

# パラメータの学習
model_lr.fit(x, y)

#予測
y_pred = model_lr.predict(x)
````

![slide35](slide/スライド35.JPG)

````python
# 描画
sns.reset_defaults()                           # 設定をリセット 
plt.scatter(x, y, c='blue')                    # 点の描画
plt.plot(x, y_pred, color='red')               # 直線の描画
plt.xlabel('Average number of rooms [RM]')     # 横軸のラベル
plt.ylabel('Price in $1000\'s [MEDV]')         # 縦軸のラベル

# グラフの表示
plt.show()   
````

````python
# 回帰直線の傾き w の表示
print('Slope w : %.3f' % model_lr.coef_)

# 回帰直線の切片 b の表示
print('Intercept b : %.3f' % model_lr.intercept_)

# 回帰直線の方程式の表示
print('y = %.3fx %.3f' % (model_lr.coef_ , model_lr.intercept_))
````

![slide36](slide/スライド36.JPG)

![slide37](slide/スライド37.JPG)

````python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

# 平均絶対誤差の表示
print('MAE : %.3f' % 
        mean_absolute_error(y, y_pred))

# 決定係数の表示
print('R^2 : %.3f' % 
        r2_score(y, y_pred))
````
