# 3-2. グラフ作成

```python
# 表示するデータ
x = [1, 2, 3]
y = [10, 1, 4]

# グラフの大きさ指定  
plt.figure(figsize = (10, 6))

# グラフの描画
plt.bar(x, y, align='center', width = 0.5)

# 棒グラフそれぞれのラベル
plt.xticks(x, ['A Class', 'B Class', 'C Class'])

# xとyのラベルを設定
plt.xlabel('Class')
plt.ylabel('Score')

# グリッドを表示
plt.grid(True)
```

```python
# データの準備
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

# X軸のデータ
x = np.arange(len(y1))

# グラフの幅
w = 0.4

# グラフの大きさ指定
plt.figure(figsize = (10, 6))

# グラフの描画。y2の方はグラフの幅の分、右にずらして描画する 
plt.bar(x, y1, color = 'blue', width = w, label = 'Math first', align = 'center')
plt.bar(x + w, y2, color='green', width = w, label = 'Math final', align = 'center')

# 凡例を最適な位置に配置
plt.legend(loc = 'best')

# 棒グラフのラベル表示, グリッド表示
plt.xticks(x + w / 2, ['Class A', 'Class B', 'Class C'])
plt.grid(True)
```

```python
# データの準備
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

# X軸
x = np.array([1, 2, 3, 4, 5]) 

# グラフの大きさ指定 
plt.figure(figsize = (10, 6))

# グラフの描画
p1 = plt.bar(x, height1, color = 'blue')
p2 = plt.bar(x, height2, bottom = height1, color='lightblue')

# 凡例を表示 
plt.legend((p1[0], p2[0]), ('Class 1', 'Class 2'))
```

```python
# 内容を表すラベルのリスト
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']

# データのリスト
sizes = [15, 30, 45, 10]

# 色のリスト
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# グラフの大きさ指定
plt.figure(figsize = (15, 6))

# グラフを表示
plt.pie(sizes,  labels = labels, colors = colors, counterclock = False,
        autopct = '%1.1f%%',  startangle = 90)

# 円を丸く描画
plt.axis('equal')
```

```python
# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)

# 散布図を描画
plt.scatter(x, y)
plt.grid(True)
```

```python
# 乱数を生成
x = np.random.rand(10)
y = np.random.rand(10)

 # グラフの大きさ指定
plt.figure(figsize = (10, 6))

# 円のサイズを作成
size = range(10)
newsize = [n*100 for n in size]#配列全体に100倍

# 散布図を描画
plt.scatter(x, y, s=newsize, c="pink",
            edgecolors="red")

plt.grid(True)
```

```python
# 平均 50, 標準偏差 10 の正規乱数を1,000件生成
x = np.random.normal(50, 10, 1000)
# yは出現回数

# ヒストグラムを出力
plt.hist(x, bins=16)

# グリッドをつける
plt.grid(True)
```

```python
# 平均 50, 標準偏差 10 の正規乱数を1,000件生成
x = np.random.normal(50, 10, 1000)

# 平均 20, 標準偏差 10 の正規乱数を1,000件生成
y = np.random.normal(20, 10, 1000)

# グラフの描画
plt.hist([x, y], bins=16, stacked=True)

# グリッドをつける
plt.grid(True)
```
