# 6-1. Seabornによる可視化

```python
# 以下のライブラリを使うので、あらかじめ読み込んでおいてください
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

# 可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

# 小数第3位まで表示
%precision 3
```

## 6-1-1. データの読み込み

```python
# データの読み込み
penguins = sns.load_dataset('penguins')
```

```python
type(penguins)
```

```python
# データの確認
penguins.head()
```

## 6-1-2. カテゴリごとのデータ分布

```python
# 種類ごとの分布を描画
sns.stripplot(x = "species", y = "bill_length_mm", data=penguins)
```

```python
# さらに性別で分ける
sns.stripplot(x = "species", y = "bill_length_mm",hue = "sex", data=penguins)
```

```python
# 箱ひげ図の描画
sns.boxplot(x = "species", y = "bill_length_mm", data=penguins)
```

```python
# バイオリン図の描画
sns.violinplot(x = "species", y = "bill_length_mm", data=penguins)
```

## 6-1-3. すべての変数を一度に比較

```python
# 散布図行列を作成
sns.pairplot(penguins)
```

```python
# ペンギンの種類によって色分けする
sns.pairplot(penguins, hue='species')
```
