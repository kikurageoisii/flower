# 3章演習 ①東京の降水量を比較してみよう

### 東京の降水量

- 1991年から2020年までの月ごとの平均降水量(mm)

|  1月  |  2月  |  3月  |  4月  |  5月  |  6月  |  7月  |  8月  |  9月  |  10月  |  11月  |  12月  |
| :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: |:----: |:----: |
|  59.7  |  56.5 |  116.0  |  133.7  |   139.7  |  167.8 |  156.2  |  154.7  |   224.9  |  234.8  | 96.3  |57.9  |

(https://www.data.jma.go.jp/obd/stats/etrn/view/nml_sfc_ym.php?prec_no=44&block_no=47662&year=&month=&day=&view=)

- 2021年の月ごとの降水量(mm)


|  1月  |  2月  |  3月  |  4月  |  5月  |  6月  |  7月  |  8月  |  9月  |  10月  |  11月  |  12月  |
| :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: | :----: |:----: |:----: |
|  43.5  |  88.5 |  173.0  |  156.0  |   99.5  |  168.5 |  310.0 |  382.5  |  222.5  | 199.5  | 93.0  | 54.5  |


(https://www.data.jma.go.jp/obd/stats/etrn/view/monthly_s1.php?prec_no=44&block_no=47662&year=2021&month=&day=&view=p1)

```python
#データの準備
y1 = np.array([59.7, 56.5, 116.0, 133.7, 139.7, 167.8, 156.2, 154.7, 224.9, 234.8, 96.3, 57.9])
y2 = np.array([43.5, 88.5, 173.0, 156.0, 99.5, 168.5, 310.0, 382.5, 222.5, 199.5, 93.0, 54.5])

# X軸のデータ
x = np.arange(len(y1))

# グラフの幅
w = 0.4

# グラフの大きさ指定(10,6)にしましょう
plt.figure(figsize = (10, 6))

# グラフの描画。y2の方はグラフの幅の分、右にずらして描画する
# y1のラベルは'Average'、y2のラベルは'2021'などわかりやすく
plt.bar(x, y1, color = 'blue', width = w, label = 'Average', align = 'center')
plt.bar(x + w, y2, color='green', width = w, label = '2021', align = 'center')

# 凡例を最適な位置に配置
plt.legend(loc = 'best')

# 棒グラフのラベル表示
# ラベルは['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(x + w / 2, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 軸ラベル
plt.xlabel('Month')
plt.ylabel('Precipitation amount (mm)')

# グラフタイトル
plt.title('Tokyo')

# グリッド表示
plt.grid(True)
```

# 3章演習 ③企業を分析してみよう

- 企業6社の財務情報

|    |  A社  |  B社  |  C社  |  D社  |  E社  |  F社  |
| :----: | :----: | :----: | :----: | :----: | :----: |:----: |
| 売上高(百万円)  |  10,264,602 |  3,336,967  |  4,476,758  |   3,014,095  | 3,586,839 |7,388,791 |
| 営業利益率(%)  |  7.19 |  4.76  | 5.63 |   4.40  |  6.11 | 4.84 |
| 時価総額(百万円)  |  6,414,556 |  2,486,212  |  3,054,394  | 1,402,448  | 3,994,100 |2,805,996 |


```python
# データラベル作成
labels = ['A', 'B', 'C', 'D', 'E', 'F']

# 売上高と営業利益率を格納
x = np.array([10264602,3336967,4476758,3014095,3586839,7388791])
y = np.array([7.19,4.76,5.63, 4.40, 6.11,4.84 ]) 

# グラフの大きさ指定(10,6)にしましょう
plt.figure(figsize = (10, 6))

# 円のサイズを作成
size = [ 6414556,2486212,3054394,1402448,3994100,2805996 ] 
#配列全体に1/10000倍
newsize = [n/10000 for n in size]

# 軸ラベル
plt.xlabel('Sales (millions)')
plt.ylabel('Operating Margin Ratio (%)')

# 散布図を描画
plt.scatter(x, y, s=newsize, c="blue", edgecolors="red")

# データ名を記述
for i, label in enumerate(labels):
    plt.text(x[i]+200000, y[i]-0.1,label)

#グラフタイトル
plt.title('Financial information')

plt.grid(True)
```
