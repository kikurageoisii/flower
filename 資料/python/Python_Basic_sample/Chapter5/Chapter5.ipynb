## 5.1.3. 要素の参照

```python
import numpy as np
# 0～1のランダムな数値より5×5の行列を作成
a = np.random.rand(5, 5)
print(a)
```

```python
# 1行3列目の要素を参照
print(a[0, 2])
```

```python
# 1行目を参照
print(a[0])
```

```python
# 3列目の要素を1次元配列で参照
print(a[:, 2])
```

```python
# 2～3行目、2～4列目を参照（表示は2×3の行列となる）
print(a[1:3, 1:4])
```

### 条件式による要素の参照

```python
import numpy as np
# 0～1のランダムな数値より4×4の行列を作成
a = np.random.rand(4, 4)
print(a)
```

```python
# 0.5より大きい数値の要素を参照
print(a[a > 0.5])
```

```python
import numpy as np
# 3×3の行列に整形
a1 = np.arange(9).reshape(3, 3)
print(a1)
```

```python
# 行列の転置
print(a1.T)
```

```python
# 要素の総和
print(a1.sum())
```

```python
# 行列同士の積
print(a1.dot(np.arange(3)))
```

## 5.2.2. グラフを描画してみる
#### quadraticfunc.py

```python
import numpy as np
from matplotlib import pyplot as plt # 慣例に従い、pltに
  
x = np.linspace(-5, 5, 30) # -5から5までの範囲を30分割した値の配列
y = np.power(x, 2)         # y = xの2乗
  
plt.plot(x, y) # x軸とy軸を描画する
plt.show()     # グラフを表示する
```

#### trigonometric.py

```python
import numpy as np
from matplotlib import pyplot as plt
  
x = np.linspace(-np.pi, np.pi) 
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.xlim(x.min(), x.max()) # x軸の描画領域を値の最小値と最大値にする
  
plt.legend() # 凡例
plt.title(r"$\sin(x)$ and $\cos(x)$") # グラフのタイトルを数式スタイルで
plt.xlabel("X-Axis") # y軸の名前
plt.ylabel("Y-Axis") # x軸の名前
plt.show()
```
