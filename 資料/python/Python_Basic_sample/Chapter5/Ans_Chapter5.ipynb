**【演習 5.1】**  
媒介変数θを用いて、
$$x=a(1+cosθ)cosθ$$
$$y=a(1+cosθ)sinθ$$

と表される曲線を、カージオイド曲線と言います。
matplotlibを用いて、この曲線を表示するプログラムを完成させてください。なお、ここでは、a=1、媒介変数θ は0≦θ≦2πとします。

```python
%matplotlib inline

import numpy as np #NumPyライブラリをインポート
from matplotlib import pyplot as plt #matplotlibライブラリのpyplotモジュールをインポート

a = 1
theta = np.linspace(0, 2*np.pi, 100) #0≦θ≦2πで要素数が100の配列を生成
x = a * (1 + np.cos(theta)) * np.cos(theta) #それぞれx,yを定義
y = a * (1 + np.cos(theta)) * np.sin(theta)
plt.plot(x,y) #x軸とy軸を描画
plt.show() #グラフの表示
```
