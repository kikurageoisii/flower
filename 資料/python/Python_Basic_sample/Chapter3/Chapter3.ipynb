### if文を用いた条件分岐

#### ifcondition.py

```python
x = 1
if x == 1:
    print("xの値は1です")
else:
    print("xの値は1以外です")
```

#### elifcondition.py

```python
x = int(input('整数を入力してください>>>'))
if x == 0:
    print('xは0です')
elif x < 0:
    print('xは負の値です')
else:
    print('xは正の値です')
```

#### and_or.py

```python
x = int(input('整数を入力してください>>>'))
if 0 <= x and x < 10:
    print('0以上、10未満です')
elif x < 0 or 10 <= x:
    print('0未満、もしくは、10以上です')
```

```python
# and_or.py中のif文の簡略版
x = int(input('整数を入力してください>>>'))
if 0 <= x < 10:
    print('0以上、10未満です')
```

#### check_country.py

```python
country = input('国名を入力してください>>>')
east_asias = ['Japan', 'China', 'North Korea', 'South Korea', 'Mongolia', 'Taiwan']
if country in east_asias:
    print('東アジアの国です')
else:
    print('東アジアの国ではありません')
```

## 3.2.3. 繰り返しの中止とスキップ

#### break_continue.py

```python
i = 0
while True:
    i += 1  # 変数iに1を足しこんで更新する
    if i % 2 == 0:
        continue  # 繰り返しの最初に戻る
    elif i > 100:
        break  # 繰り返しを終了する
    print("i = {}".format(i))

print("100より大きくなったので終了しました")
```

## 3.4.2. ユーザー定義クラスの作成
#### simpleclass.py

```python
class Simple:
    def __init__(self, i='デフォルト値'):
        print('__init__メソッドが呼び出されました')
        self.instance_attr = i
        
    def instance_print(self):
        print('インスタンス変数：' + str(self.instance_attr))
        
# クラスを使った処理
simple_instance = Simple('test')  # インスタンスの作成
simple_instance.instance_print()
simple_instance.instance_attr = '上書き'  # インスタンス変数に値を割り当て
simple_instance.instance_print()
print(simple_instance.instance_attr)
```

## 3.4.4. クラス変数とインスタンス変数
#### classattr.py

```python
class Simple:
    class_attr = 'クラス変数'
    def __init__(self, i):
        self.instance_attr = i
    
s1 = Simple('s1')
s2 = Simple('s2')
print(Simple.class_attr)
print(s1.instance_attr)
print(s2.instance_attr)
```
