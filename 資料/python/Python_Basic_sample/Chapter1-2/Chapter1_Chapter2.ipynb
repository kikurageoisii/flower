### (ii) リストの要素の参照

```python
x = [2, 3, -5, 11, 7]
print(x[1])
```

```python
print(x[-1])
```

#### calcset.py

```python
a = {1, 2, 4, 6, 8}
b = {1, 3, 5, 7, 9}
  
print(a | b)  # aとbの和集合
print(a & b)  # aとbの積集合
print(a - b)  # aとbの差集合
```

#### subset.py

```python
a = {1, 2, 3, 4, 5}
b = {2, 3}
  
print({4} <= a)  # 4がaに含まれているか
print(a <= b)  # aがbの部分集合であるか
print(a >= b)  # bがaの部分集合であるか
```

```python
# 要素間に順序は存在しない
s = {1, 2, 3}
print(s[1])
```

## 2.2.4. 辞書

```python
d = {'apple': 100, 'orange': 75}
print(d)
```

```python
d = {['Hello', 'World']: 1}
```

```python
# あるキーに対する値の確認
d = {'apple': 100, 'orange': 75}
print(d['apple'])
```

```python
# 値の変更
d = {'apple': 100, 'orange': 75}
d['apple'] = 80
print(d)
```

```python
d={"apple": 100, "orange": 75}
print(d["banana"])
```

```python
# 要素の追加
d={"apple": 100, "orange": 75}
d["banana"] = 100
print(d)
```

```python
# 要素の削除①
d={"apple": 100, "banana": 100, "orange": 75}
d.pop("banana")
print(d)
```

```python
# 要素の削除②
del d["apple"]
print(d)
```

```python
# 要素の存在確認
d={"apple": 100,"banana": 100,"orange": 75}
print("apple" in d)
print(100 in d)
```

```python
# キーの一覧を取得
d={"apple": 100,"banana": 100,"orange": 75}
print(d.keys())
```

```python
# 値の一覧を取得
print(d.values())
```

```python
# 値の存在確認
print(100 in d.values())
```
