## 4.1.2. ファイルの読み込み
#### fread.py

```python
myfile = open('test.txt', mode='r')

data = myfile.read()
print(data)

myfile.close()
```

#### freadlines.py

```python
myfile = open('test.txt', mode='r')

datalist  = myfile.readlines()
print("datalist:{}".format(datalist))

for data in datalist:
    print(data)
    
myfile.close()
```

#### freadline.py

```python
myfile = open('test.txt', mode='r')

data = myfile.readline()
print(data)

myfile.close()
```

```python
# 出力
def print_test_txt():
    myfile = open('test.txt', mode='r')

    data = myfile.read()
    print(data)

    myfile.close()
    
```

## 4.1.3. ファイルの書き込み(上書き : w モード)
#### fwrite_w.py

```python
myfile = open('test.txt', mode='w')

myfile.write('Meiji University Hello World!!\n')

myfile.close()

# 出力
print_test_txt()
```

#### fwritelines_w.py

```python
myfile = open('test.txt', mode='w')

datalist = ['Hello World!!\n', 'Meiji\n', 'University\n']
myfile.writelines(datalist)

myfile.close()

# 出力
print_test_txt()
```

```python
# 復元用
def restore_test_txt():
    myfile = open('test.txt', mode='w')

    datalist = ['Hello World !!\n', '\n', 'Meiji University']
    myfile.writelines(datalist)

    myfile.close()
    
```

## 4.1.4. ファイルの書き込み(追記 : a モード)
#### fwrite_a.py

```python
# 復元
restore_test_txt()

myfile = open('test.txt', mode='a')

myfile.write('Meiji University Hello World!!\n')

myfile.close()

# 出力
print_test_txt()
```

#### fwriitelines_a.py

```python
# 復元
restore_test_txt()

myfile = open('test.txt', mode='a')

datalist = ['Hello World!!\n', 'Meiji\n', 'University\n']
myfile.writelines(datalist)

myfile.close()

# 出力
print_test_txt()
```
