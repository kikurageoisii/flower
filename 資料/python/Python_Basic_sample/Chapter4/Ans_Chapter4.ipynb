**【演習 4.1】**  
外部ファイルとして、fibonacci.txtというファイルを用意しました。
 
このテキストを読み込んで、各数字に対して偶数か奇数かを判定するプログラムを完成させてください。

```python
f = open("fibonacci.txt", "r") # ファイルを読み込みようにオープン
text = f.readline()            # ファイルの1行目を読み込み
data = text.split(",")         # splitメソッドで各要素に切り分け
for num in data:               # 偶奇判定
    if int(num)%2==0:
        print(num+"は偶数")
    elif int(num)%2!=0:
        print(num+"は奇数")
f.close()                      # 読み込んでいたファイルをクローズ
```
