# 解答例

## ライブラリのインポート

```python
# 外部ライブラリのインポート
import numpy as np 
from matplotlib import pyplot as plt
```

```python
#自作モジュールのインポート
from utils.ex_class import Person
from utils.ex_function import get_float_list
```

## データの読み込み
ここでは演習4.2.2で使用したget_float_list関数を利用します。

get_float_listはテキストファイルを読み込み、float型へ変換されたリストを返す関数です。

```python
#　体重・身長データの読み込み
weights = get_float_list("data/weights.txt")
heights = get_float_list("data/heights.txt")
print("weights:{}".format(weights))
print("heights:{}".format(heights))
```

## Personクラスを利用したBMI計算

```python
# Personクラスを用いたBMI計算
person = Person(name="Taro", age=20)
person.set_weight(weights)
person.set_height(heights)
bmi = person.calc_BMI("transition")
x = np.linspace(1, 12, 12)   # x = np.arange(1, 13) でも可
y = bmi


# BMIをグラフに描画
plt.plot(x, y, label="BMI")

# 【オプション】　基準値をグラフ内に描画
x = np.linspace(1,12,12)
y = [25]*12
plt.plot(x,y, label="standard value")

plt.title("name : {},   age : {}".format(person.name, person.age)) # タイトルの設定
plt.xlabel("month")                      # x軸の名前の設定
plt.ylabel("BMI")                        # y軸の名前の設定
plt.xticks(np.linspace(1, 12, 12))       # x軸の目盛りの設定
plt.grid()                               # gridを追加
plt.legend()
```
